# Implementing Search in the Heritage Connector

```python
%load_ext autoreload
%autoreload 2

import sys
sys.path.append("../..")

from heritageconnector.disambiguation import search
```

### Wikidata search

```python
wd_search = search.wikidata_text_search()

property_filters = {
    "P106": "Q39631", # occupation: physician 
}

# without property filter
wd_search.run_search("John Snow", instanceof_filter="Q5", include_class_tree=True)
```

```python
# with property filter
wd_search.run_search("John Snow", instanceof_filter="Q5", include_class_tree=True, property_filters=property_filters)
```

### Wikipedia search

```python
from fuzzywuzzy import fuzz
print( fuzz.token_set_ratio("Nokia N-Gage", "Nokia 'N-Gage' mobile telephone, 2004-2007") )
print( fuzz.token_set_ratio("John Snow", "Snow, J") )
print( fuzz.token_set_ratio("John Snow", "Brittany Snow") )
print( fuzz.token_set_ratio("John Snow", "Snow (musician)") )
print( fuzz.token_set_ratio("John Snow", "Danielle Panabaker") )
```

```python
wk_search = search.wikipedia_text_search()
wk_search.run_search("John Snow", instanceof_filter="Q5", limit=100)
```

```python
# with property filter
wd_search.run_search("John Snow", instanceof_filter="Q5", include_class_tree=True, property_filters=property_filters)
```
